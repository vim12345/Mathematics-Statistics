```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
# Create synthetic linear data
X, y = make_regression(n_samples=200, n_features=1, noise=15.0, random_state=42)
# Add bias column
X_b = np.c_[np
            .ones((X.shape[0], 1)), X]
# Loss: Mean Squared Error
def compute_loss(theta, X, y):
  preds = X.dot(theta)
  return np.mean((preds - y)**2) / 2

# Gradient descent
def gradient_descent(X, y, lr=0.01, n_iters=1000):
  theta = np.zeros(X.shape[1])
  n = X.shape[0]
  losses = []
  for i in range(n_iters):
    preds = X.dot(theta)
    grad = (1/n) * X.T.dot(preds - y)
    theta = theta - lr * grad
    losses.append(compute_loss(theta, X, y))
  return theta, losses
theta_opt, losses = gradient_descent(X_b, y, lr=0.1, n_iters=500)
print("Theta (bias, weight):", theta_opt)
plt.plot(losses)
plt.title("MSE during gradient descent")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.show()
```
