Calculate mean, median, variance by hand on a small dataset.

```python
import numpy as np
import matplotlib.pyplot as plt

# Part 1: Gradient Descent Simulation and Plot
def gradient_descent(starting_x, learning_rate, num_iterations):
    x = starting_x
    x_history = [x]
    f_history = [x**2]  # f(x) = x^2

    for _ in range(num_iterations):
        gradient = 2 * x  # Derivative of x^2 is 2x
        x -= learning_rate * gradient
        x_history.append(x)
        f_history.append(x**2)

    return x_history, f_history

# Parameters
starting_x = 10
learning_rate = 0.1
num_iterations = 10

x_history, f_history = gradient_descent(starting_x, learning_rate, num_iterations)

# Plot the loss curve and descent path
x_vals = np.linspace(-11, 11, 100)
f_vals = x_vals**2

plt.plot(x_vals, f_vals, label='f(x) = x^2')
plt.plot(x_history, f_history, 'ro-', label='Gradient Descent Path')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent on f(x) = x^2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Part 2: Statistics on Dataset
dataset = np.array([1, 3, 5, 2, 4])

# Mean
mean = np.mean(dataset)
print(f"Mean: {mean}")

# median
median = np.median(dataset)
print(f"Median: {median}")

# variance

population_variance = np.var(dataset, ddof=0)
sample_variance = np.var(dataset, ddof=1)
print(f"Population Varinace: {population_variance}")
print(f"Sample Variance: {sample_variance}")
```
